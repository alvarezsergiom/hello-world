## def creates customized function

```python
# defining a funtion named hello
def hello(to):
    print("hello,", to)
#ask for a name
name = input("What's your name? ")
#print hello function
hello(name)
```

```python
# defining a funtion named hello
def hello(to = "world"):
    print("hello,", to)
#ask for a name
hello()
name = input("What's your name? ")
#print hello function
hello(name)
```
